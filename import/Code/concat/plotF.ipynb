```python
import matplotlib.pyplot as plt
from matplotlib.dates import MonthLocator, DateFormatter

# Assuming 'Month_Day' is a datetime column
plt.figure(figsize=(15,8))
plt.plot(df['Month_Day'], df['2023'])
plt.xlabel('Month')
plt.ylabel('2023 quantities')
plt.title('REGRESSION - 2023(Quantities) vs Month_Day')

# Set x-axis to show months
plt.gca().xaxis.set_major_locator(MonthLocator())
plt.gca().xaxis.set_major_formatter(DateFormatter("%b"))

plt.xlim(df['Month_Day'].min(), df['Month_Day'].max())
plt.show()
```
